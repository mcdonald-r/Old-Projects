```python
class Goat:
    def __init__(self, n, a, c):
        self.__name = n
        self.__age = a
        self.__color = c
    
    def birthday(self):
        self.__age += 1
        print(f'Happy Birthday, {self.__name}! You\'re now {self.__age}!')
    
    def __str__(self):
        return f'{self.__name} ({self.__age}) [{self.__color}]'
    
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    def get_color(self):
        return self.__color
    
    def set_name(self, n):
        self.__name = n;
    
    def set_age(self, a):
        self.__age = a
    
    def set_color(self, c):
        self.__color = c

def main():
    infile = open('goats.txt')
    trip = []  
    name = infile.readline().rstrip()
    while name != '':
        age = int(infile.readline().rstrip())
        color = infile.readline().rstrip()
        trip.append(Goat(name, age, color))
        name = infile.readline().rstrip()
        
    value = 0
    for goat in trip:
        while value != 9:
            i = trip.index(goat)
            value = menu(trip, i)
            if value == 1:
                change_name(trip, i)
            elif value == 2:
                change_age(trip, i)
            elif value == 3:
                change_color(trip, i)
            elif value == 4:
                i = (i + 1) % len(trip)
                value = menu(trip, i)
            elif value == 5:
                i = (i - 1) % len(trip)
                value = menu(trip, i)
            elif value == 6:
                add_goat(trip)
            elif value == 7:
                remove_goat(trip, i)
            elif value == 8:
                print_trip(trip, i)
        
def menu(trip, index):
        print('*** GOAT INFORMATION MANAGEMENT SYSTEM ***')
        print(f'Current Goat: {trip[index]}')
        print('**********************************************')
        print('*** Select: [1] Change name')
        print('***         [2] Change age')
        print('***         [3] Change color')
        print('***         [4] Next goat')
        print('***         [5] Prev goat')
        print('***         [6] Add a goat')
        print('***         [7] Remove current goat')
        print('***         [8] Print trip')
        print('***         [9] Quit')
        choice = int(input('*** Choice: '))
        print('')
        while choice not in range(1,10):
            choice = int(input('Error: Must enter integer 1 - 9 *** Choice: '))
        return choice
    
def change_name(trip, index):
    new_name = input('Enter the new goat name: ')
    print('')
    trip[index].set_name(new_name)
    
def change_age(trip, index):
    new_age = int(input('Enter the new goat age: '))
    print('')
    trip[index].set_age(new_age)
    
def change_color(trip, index):
    new_color = input('Enter the new goat color: ')
    print('')
    trip[index].set_color(new_color)
      
def add_goat(trip):
    name = input('Enter the new goat name: ')
    age = int(input('Enter the new goat age: '))
    color = input('Enter the new goat color: ')
    print('')
    trip.append(Goat(name, age, color))
    
def remove_goat(trip, index):
    sure = input('Removing goat... are you sure? Y/N: ')
    while sure.upper() != 'Y' and sure.upper() != 'N':
        sure = input('Error: enter \'Y\' or \'N\' Removing goat... are you sure? Y/N: ')
    if sure.upper() == 'Y':
        if len(trip) == 0:
            pass
        else:
            del trip[index]
    else:
        pass
    
def print_trip(trip, index):
    for goat in trip:
        print(goat)
    print('')
    
main()
```
